```python
import random
import torch
from d2l import torch as d2l
```

```python
# 模拟创造一个数据集  已知w,b及函数，模拟出一个x，y的数据集，再根据数据集训练得到w,b和真实值比较，验证准确性
def synthetic_data(w, b, num_examples):
    """生成 y = Xw + b + 噪声"""
    x = torch.normal(0,1,(num_examples,len(w)))   # 均值为0，方差为1
    y = torch.matmul(x, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return x, y.reshape(-1,1)

true_w = torch.tensor([2,-3.4])
true_b = 4.2
features, label = synthetic_data(true_w,true_b,1000)
```

```python
# 接收一个批量数据及标签
def data_iter(batch_size, features, label):
    num_examples = len(features)
    indices = list(range(num_examples))       # 生成和feature个数一样的索引
    random.shuffle(indices)                   # 打乱顺序
    for i in range(0,num_examples,batch_size):    
        batch_indices = torch.tensor(
        indices[i:min(i + batch_size,num_examples)])
        yield features[batch_indices],label[batch_indices]
batch_size = 5
for x,y in data_iter(batch_size,features,label):
    print(x,'\n',y)
    break
```

```python
# 定义初始化模型参数
w = torch.normal(0,0.01,size = (2,1),requires_grad=True)
b = torch.zeros(1,requires_grad=True)
```

```python
# 定义模型  线性回归模型
def linreg(x, w, b):
    return torch.matmul(x, w)+b 
```

```python
# 定义损失函数
def squared_loss(y_hat, y):
    """均方损失"""
    return (y_hat - y.reshape(y_hat.shape))**2 /2
```

```python
# 定义优化算法
def sgd(params, lr, batch_size):
    """小批量梯度随机下降。"""
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()
```

```python
# 训练过程
lr = 0.001
num_epochs = 30
net = linreg
loss = squared_loss

for epoch in range(num_epochs):
    for x, y in data_iter(batch_size, features, label):
        l = loss(net(x,w,b),y)
        l.sum().backward()
        sgd([w,b],lr,batch_size)   # 使用参数梯度进行更新
    with torch.no_grad():
        train_l = loss(net(features,w,b),label)
        print(f'epoch{epoch + 1},loss {float(train_l.mean()):f}')
```
