```python
# https://zhuanlan.zhihu.com/p/100064394?utm_source=wechat_session&utm_medium=social&utm_oi=73846707191808&utm_campaign=shareopn
# map是针对单列，apply是针对多列，applymap是针对全部元素
import pandas as pd
import numpy as np
```

```python
boolean=[True,False]
gender=["男","女"]
color=["white","black","yellow"]
data=pd.DataFrame({
    "height":np.random.randint(150,190,100),
    "weight":np.random.randint(40,90,100),
    "smoker":[boolean[x] for x in np.random.randint(0,2,100)],
    "gender":[gender[x] for x in np.random.randint(0,2,100)],
    "age":np.random.randint(15,90,100),
    "color":[color[x] for x in np.random.randint(0,len(color),100) ]
}
)
```

```python
data
```

```python
# map
# 使用字典映射
data['gender2'] = data['gender'].map({'男':1,'女':0} )
```

```python
# map使用方法2 函数
def gender_map(x):
    gender = 1 if x == "男" else 0
    return gender

# 调用函数传入函数名
data['map+函数'] = data['gender'].map(gender_map)
data
```

```python
# apply  apply方法的作用原理和map方法类似，区别在于apply能够传入功能更为复杂的函
def apply_age(x,bias,bias2):
    return x+bias+ bias2
#以元组的方式传入额外的参数
data["age2"] = data["age"].apply(apply_age, args=(-3,1,))
data
```

```python
data['gender2'] = data['gender'].apply(lambda x: 1 if x == '男' else 0 )
```

```python
data
```

```python
# 对datafeame的处理
# axis=0代表操作对列columns进行，axis=1代表操作对行row进行

# 沿0轴对data求和
data[['height','weight']].apply(np.sum,axis=0)
```

```python
data[['height','weight']].apply(lambda x: x*x)
```

```python
def BMI(series):
    weight = series["weight"]
    height = series["height"]/100
    BMI = weight/height**2
    return BMI

data["BMI"] = data.apply(BMI,axis=1)
```
